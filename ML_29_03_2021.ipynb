# Perform dataset handling operations using Pandas

```python
#importing library first
import pandas as pd
```

```python
#importing the dataset and displaying it
pokemon = pd.read_csv("C:\\Prerna-Misc\\SEM - 6\\ML\\pokemon.csv")
print(pokemon)
```

```python
#find the size of the dataset
print(pokemon.size)
```

```python
#find the shape of the dataset
print(pokemon.shape)
```

```python
#find the data type of all the tuples in the dataset
print(pokemon.dtypes)
```

```python
#using head and tail function
#returns five first and last rows respectively by default
#a number passed as parameter can change the number of the rows to be printed
# [always followed by 'm' rows x 'n' columns in the output]

print(pokemon.head())

print(pokemon.tail(8))
```

```python
#to see how many row have the column type1 equal to bug
pokemon.loc[(pokemon.type1 == 'bug')]
```

```python
#to find rows where type1 is either steel or bug
pokemon.loc[(pokemon['type1']=='steel') | (pokemon['type1']=='bug')]
```

```python
#To retrieve value at particular position
#Read specific position
print(pokemon.iloc[10,25])
```

```python
#Read data of every 6th row from 1 to 25
print(pokemon.iloc[1:25:6])
```

```python
#To create a new column
#hp, attack and defense are existing columns in the dataset
pokemon['total'] = pokemon['hp'] + pokemon['attack'] + pokemon['defense']
```

```python
#To check if the above new column has been created or not
print(pokemon.total.head())
```

```python
#Drop a column
#If you don't want to alter the existing data then create a new dataset
pokemon1 = pokemon.drop(columns=["total"])
```

```python
#Check if the column has been dropped
print(pokemon1.total)
```

```python
#isnull function checks for missing values and returns 1
#we sum 1s to get count of misisng values by columns
pokemon.isnull().sum()
```

```python
#Fill missing values with column mean
pokemon = pokemon.fillna(pokemon.mean())
```

```python
#Fill missing values with 0
pokemon = pokemon.fillna(0)
```

```python
# Check if missing values have been subsituted or not
pokemon.isnull().sum()
```
